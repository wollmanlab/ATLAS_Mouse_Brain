```python
charpair = {
    'A': "T",
    'T': "A",
    'C': "G",
    'G': "C",
}

def rev_comp(seq):
    rseq = ""
    for char in seq[::-1]:
        rseq += charpair[char]
    return rseq 
```

```python
newname = "RS0639"
newseq = "TATCTCATCAATCCCACACT"
print(newseq)
print(rev_comp(newseq))
```
