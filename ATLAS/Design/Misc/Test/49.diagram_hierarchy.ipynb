```python
import numpy as np
from dredFISH.Utils.__init__plots import *
```

```python
# a two-level hierarchy
nt = 5**2
mat = np.random.randn(nt, 2)
mat_a = 10*np.random.randn(nt//5, 2)
mat += np.repeat(mat_a, 5, axis=0)

plt.scatter(mat[:,0], mat[:,1], label='L(-1)')
plt.scatter(mat_a[:,0], mat_a[:,1], label="L(-2)")
plt.legend(bbox_to_anchor=(1,1))
```

```python
# a three-level hierarchy

n_lvl = 2
n_perlvl = 5 
scale = 10 
nt = n_perlvl**n_lvl

# start
_n = nt
_s = 1
tsr = np.zeros((nt,2,n_lvl))

# mat = np.zeros((nt,2))
for i in range(n_lvl):
    _mat = _s*np.random.randn(_n, 2)
    _mat_expand = np.repeat(_mat, nt//_n, axis=0)
    # mat += _mat_expand 
    
    for j in range(i+1):
        tsr[:,:,j] += _mat_expand
    
    _n = _n//n_perlvl
    _s = _s*scale
```

```python
fig, ax = plt.subplots(figsize=(8,6))
for i in range(n_lvl):
    ax.scatter(tsr[:,0,i], tsr[:,1,i], 
               # s=5*(i+1), 
               s=10,
               label=f'L(-{i})')
ax.legend(bbox_to_anchor=(1,1))
ax.grid(False)
ax.set_aspect('equal')
```
