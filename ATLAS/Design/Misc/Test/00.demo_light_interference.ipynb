```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0, 2*np.pi, 100)
```

```python
# check trig
# plt.plot(a, 1+np.cos(a), '-.')
plt.plot(a, 2*np.cos(a/2), '-.')
plt.plot(a, np.sin(a)/np.sin(a/2), '-.')
```

```python
theta = np.linspace(-np.pi/2, np.pi/2, 500)
d = 1 # measured in \lambda
ds = [1, 2, 5]
fig, ax = plt.subplots()
for d in ds:
    a = (2*np.pi*d)*np.sin(theta)
    plt.plot(theta*180/np.pi, (2*np.cos(a/2))**2, '-', label=f"d={d} lmd")
plt.xlim([-90, 90])
plt.ylim([-0.1,4.1])
plt.xticks(np.linspace(-90, 90, 13))
plt.xlabel('degree')
plt.xlabel('intensity')
plt.axvline(0, linestyle='--', color='gray')
plt.legend(bbox_to_anchor=(1,1))
```

```python
ds = [0.01, 0.1]
fig, ax = plt.subplots()
for d in ds:
    a = (2*np.pi*d)*np.sin(theta)
    plt.plot(theta*180/np.pi, (2*np.cos(a/2))**2, '-', label=f"d={d} lmd")
plt.xlim([-90, 90])
plt.xticks(np.linspace(-90, 90, 13))
plt.xlabel('degree')
plt.xlabel('intensity')
plt.ylim([-0.1,4.1])
plt.axvline(0, linestyle='--', color='gray')
plt.legend(bbox_to_anchor=(1,1))
```

```python
theta = np.linspace(-np.pi/2, np.pi/2, 500)
d = 1 # measured in \lambda
ds = [10.5]
fig, ax = plt.subplots()
for d in ds:
    a = (2*np.pi*d)*np.sin(theta)
    plt.plot(theta*180/np.pi, (2*np.cos(a/2))**2, '-', label=f"d={d} lmd")
plt.xlim([-90, 90])
plt.ylim([-0.1,4.1])
plt.xticks(np.linspace(-90, 90, 13))
plt.xlabel('degree')
plt.xlabel('intensity')
plt.axvline(0, linestyle='--', color='gray')
plt.legend(bbox_to_anchor=(1,1))
```

```python
theta = np.linspace(-np.pi/2, np.pi/2, 500)
fig, ax = plt.subplots()
Ns = [10, 100, 1000]
for N in Ns:
    d = 1/N
    a = (2*np.pi*d)*np.sin(theta)
    plt.plot(theta*180/np.pi, ((np.sin(N*a/2)/np.sin(a/2))**2)/N**2, '-', label=f"d={d} lmd")
plt.xlim([-90, 90])
plt.ylim([-0.1,1.1])
plt.xticks(np.linspace(-90, 90, 13))
plt.xlabel('degree')
plt.xlabel('intensity')
plt.axvline(0, linestyle='--', color='gray')
plt.legend(bbox_to_anchor=(1,1))
```

```python
theta = np.linspace(-np.pi/2, np.pi/2, 500)
Ds = [0.01, 0.1, 1,10, 100] # D = 10 # in lambda 
for D in Ds:
    A = 2*np.pi*D*np.sin(theta)
    i = np.power(np.sin(A/2)/(A/2), 2)
    plt.plot(theta*180/np.pi, i, label=f"D={D}")
plt.xticks(np.linspace(-90, 90, 13))
plt.xlim([-90, 90])
plt.legend()
plt.show()
```
