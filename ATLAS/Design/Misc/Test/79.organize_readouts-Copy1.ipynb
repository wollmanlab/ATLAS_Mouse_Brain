```python
import numpy as np
import pandas as pd
import os
```

```python
charpair = {
    'A': "T",
    'T': "A",
    'C': "G",
    'G': "C",
}

def rev_comp(seq):
    rseq = ""
    for char in seq[::-1]:
        rseq += charpair[char]
    return rseq 
```

# load old

```python
ddir = "/bigstore/GeneralStorage/fangming/projects/dredfish/res_seq"
f = os.path.join(ddir, "readout_25bits_Nov16_2022.csv")
df = pd.read_csv(f, index_col=0)
df
```

# add new

```python
ddir = "/bigstore/GeneralStorage/fangming/projects/dredfish/res_seq"
f = os.path.join(ddir, "readout_Fang_etal_2022_Science.csv")
df2 = pd.read_csv(f, index_col=0)
df2 = df2.iloc[20:24]  # previous analysis
df2.columns = ['name', 'seq_readout']
df2['seq_encode'] = df2['seq_readout'].apply(rev_comp)
df2['name'] = np.char.add(df2['name'].values.astype(str), '-Fang22')

df2
```

```python
dfall = pd.concat([df, df2], axis=0, ignore_index=True)
dfall
```

# save

```python
fout = os.path.join(ddir, "readout_29bits_Nov17_2022.csv")
dfall.to_csv(fout)
```
