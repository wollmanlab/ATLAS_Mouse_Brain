```python
import nupack
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
f = "/greendata/GeneralStorage/fangming/projects/dredfish/res_seq/readout_29bits_Nov17_2022.csv"
df = pd.read_csv(f, index_col=0)
seqs = df['seq_readout'][:24]
seqs_enc = df['seq_encode'][:24]
seqs
```

```python
%%time
# specify strands
strands =  [nupack.Strand(seq, name=f"e{i}") for i, seq in enumerate(seqs_enc)] #[:20]
strands += [nupack.Strand(seq, name=f"r{i}") for i, seq in enumerate(seqs)][0:1]

# specify tubes
t1 = nupack.Tube(strands={strand: 1e-8 for strand in strands}, 
                 complexes=nupack.SetSpec(max_size=2), 
                 name='t1')

# analyze tubes
model1 = nupack.Model(material='dna', 
                      celsius=37,
                      sodium=1,
                     )
tube_results = nupack.tube_analysis(tubes=[t1], model=model1)
# tube_results
```

```python
tube_results
```

```python
conc = pd.Series({key.name.strip("()"): item for key, item in 
        tube_results['t1'].complex_concentrations.items()
       })
conc
```

```python
plt.plot(np.sort(conc))
plt.yscale('log')
plt.xlabel('Complex')
plt.ylabel('Concentration (M)')
```

```python
def fraction_of_signal(conc, signal='r0+e0'):
    """
    """
    conc_sub = conc.filter(regex=r'\+').filter(regex=r'r')
    frac = conc_sub.loc[signal]/conc_sub.sum()
    return frac
```
